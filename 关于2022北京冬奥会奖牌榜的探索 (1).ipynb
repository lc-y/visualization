```python
tmp = df_t.query('Â•ñÁâåÁ±ªÂûã == "ÈáëÁâå"')['Ëé∑Â•ñÊó∂Èó¥'].value_counts().reset_index()
tmp.columns = ['Êó•Êúü','‰∏™Êï∞']
tmp = tmp.sort_values('Êó•Êúü')
for i in range(1,len(tmp)):
    tmp.iloc[i,1] = tmp.iloc[i-1,1] + tmp.iloc[i,1]
```

```python
def cal_everyday(type_):
    tmp = df_t[df_t['Â•ñÁâåÁ±ªÂûã'] == type_]['Ëé∑Â•ñÊó∂Èó¥'].value_counts().reset_index()
    tmp.columns = ['Êó•Êúü',type_+'ÊÄªÊï∞']
    tmp = tmp.sort_values('Êó•Êúü')
    for i in range(1,len(tmp)):
        tmp.iloc[i,1] = tmp.iloc[i-1,1] + tmp.iloc[i,1]
    return tmp
```

```python
tmp = pd.merge(cal_everyday("ÈáëÁâå"),pd.merge(cal_everyday("Èì∂Áâå"),cal_everyday("ÈìúÁâå"),how='outer'),how='outer').sort_values('Êó•Êúü')
tmp.fillna(method='ffill',inplace=True)
tmp.fillna(0,inplace=True)
```

```python
line_style = {
    'normal': {
        'width': 5,
        'shadowColor': 'rgba(155, 18, 184, .3)',
        'shadowBlur': 10,
        'shadowOffsetY': 15,
        'shadowOffsetX': 10,
        'curve': 0.5
    }
}
```

```python
from pyecharts.commons.utils import JsCode
color_js = """
            new echarts.graphic.LinearGradient(
                                0,
                                0,
                                0,
                                1,
                                [{offset: 0, color: '#FFEDBC'},
                                 {offset: 1, color: '#ED4264'}],
                                false)
           """
```

```python
tl = Timeline(init_opts=opts.InitOpts(theme='light',width='980px'))
tl.add_schema(is_auto_play=True, is_loop_play=True, 
                    is_timeline_show=False, play_interval=1500)
values = []
values_s = []
values_b = []
for i in tmp['Êó•Êúü']:
    values.append(tmp[tmp['Êó•Êúü'] == i]['ÈáëÁâåÊÄªÊï∞'].values.tolist()[0])
    values_s.append(tmp[tmp['Êó•Êúü'] == i]['Èì∂ÁâåÊÄªÊï∞'].values.tolist()[0])
    values_b.append(tmp[tmp['Êó•Êúü'] == i]['ÈìúÁâåÊÄªÊï∞'].values.tolist()[0])
    line = (
        Line(init_opts=opts.InitOpts(width='980px',height='600px',bg_color=JsCode(color_js),))
        .add_xaxis(tmp['Êó•Êúü'].tolist())
        .add_yaxis(
            'Gold_Sum',
            values,
            is_smooth=True,
            linestyle_opts=line_style
            )
        .add_yaxis(
            'Silver_Sum',
            values_s,
            is_smooth=True,
            linestyle_opts=line_style
            )
        .add_yaxis(
            'Bronzes_Sum',
            values_b,
            is_smooth=True,
            linestyle_opts=line_style
            )
        .set_global_opts(
             title_opts=opts.TitleOpts(
                title="ÊàëÂõΩÈáë„ÄÅÈì∂„ÄÅÈìúÁâåÊÄªÈáèÊó•ÂèòÂåñÁä∂ÂÜµ",
                subtitle='Êï∞ÊçÆÊù•Ê∫êÔºöÁôæÂ∫¶‰ΩìËÇ≤',
                pos_right='center',
                pos_top='2%'
                ),
            legend_opts=opts.LegendOpts(
                is_show=True,
                legend_icon='circle',
                pos_right='10%',
                pos_top='7%'
                ),
            xaxis_opts=opts.AxisOpts(
                axislabel_opts={'rotate':90}
            ),
            yaxis_opts=opts.AxisOpts(
                max_=15,
                axisline_opts=opts.AxisLineOpts(
                    is_show=False
                ),
                splitline_opts=opts.SplitLineOpts(
                    is_show=True
                )
            ),
        )
    )
    tl.add(line,'{}'.format(i))
tl.render_notebook()
```
