```python
df1['国旗'] = df1['国旗'].map(lambda x:'image://'+x)
icon_data = dict(zip(df1['国家'], df1['国旗']))
icons = []
for icon,count in zip(df1['国家'],df1['总数']):
    icons.append(dict(name=icon, value=count, symbol=icon_data[icon]))
```

```python
p1 = (
    PictorialBar(init_opts=opts.InitOpts(height='1300px'))
    .add_xaxis(df1['国家'].tolist())
    .add_yaxis("",
        icons,
        label_opts=opts.LabelOpts(
            is_show=False
            ),
        #category_gap='10%',
        #symbol_pos='start',
        symbol_size=30,
        is_symbol_clip=False,)
    .set_global_opts(
        xaxis_opts=opts.AxisOpts(
            is_show=False
            ),
        yaxis_opts=opts.AxisOpts(
            is_show=True,
            is_scale=True,
            axistick_opts=opts.AxisTickOpts(
                is_show=False
                ),
            axislabel_opts=opts.LabelOpts(
                font_size=12,
                font_family='Courier New',
                color='#black',
                font_weight=700,
                margin=10
                ),
            splitline_opts=opts.SplitLineOpts(
                is_show=False, 
                linestyle_opts=opts.LineStyleOpts(
                    type_='dashed'
                    )
                ),
            axisline_opts=opts.AxisLineOpts(
                is_show=False,
                linestyle_opts=opts.LineStyleOpts(
                    width=2, 
                    color='#DB7093'
                )
            )
        ),
    )
    .reversal_axis()
)
```

```python
bar = (Bar(init_opts=opts.InitOpts(height='1300px',theme='light'))
    .add_xaxis(df1['国家'].tolist())
    .add_yaxis('金牌数',df1['金牌'].tolist(),stack=1,)
    .add_yaxis('银牌数',df1['银牌'].tolist(),stack=1,)
    .add_yaxis('铜牌数',df1['铜牌'].tolist(),stack=1,)
    .set_series_opts(
        label_opts=opts.LabelOpts(
            is_show=False,
        ),
        itemstyle_opts=opts.ItemStyleOpts(
            border_color='#5C3719',
            border_width=2
        )
    )
    .set_global_opts(
        title_opts=opts.TitleOpts(
            title='2022北京冬奥会奖牌榜奖牌组成明细',
            subtitle='数据来源：百度体育',
        ),
        legend_opts=opts.LegendOpts(
            pos_top='2.3%',
            pos_left='15%'
        ),
        xaxis_opts=opts.AxisOpts(
            is_show=False,
            max_=int(df1['总数'].max())
        ),
        yaxis_opts=opts.AxisOpts(
            is_show=False
        ),
        tooltip_opts=opts.TooltipOpts(
            is_show=True,
            trigger='axis',
            trigger_on='mousemove|clike',
            axis_pointer_type='shadow',
        )
    )
    .reversal_axis()
)
```

```python
grid = Grid(init_opts=opts.InitOpts(height='1300px'))
grid.add(p1, is_control_axis_index=False, grid_opts=opts.GridOpts(pos_right='70%'))
grid.add(bar, is_control_axis_index=False, grid_opts=opts.GridOpts(pos_left='15%', pos_right='10%'))
grid.render_notebook()
```

```python
df_t = df2.groupby(['国家','大项项目']).agg({'小项项目':'count'}).reset_index()
df_t.columns = ['国家','大项项目','数量']
```

```python
data = []
item = []
for idx, row in df_t.iterrows():
    if row['国家'] in item:
        data[-1]['children'].append(dict(name=row['大项项目'], value=row['数量']))
    else:
        data.append(dict(name=row['国家'], children=[dict(name=row['大项项目'], value=row['数量'])]))
    item.append(row['国家'])
```

```python
tree = TreeMap(    
    init_opts=opts.InitOpts(
        theme='light',
        width='1000px',
        height='600px',
        ))
tree.add(
    "奖牌数量",
    data,
    leaf_depth=1,
    roam=False,
    label_opts=opts.LabelOpts(position="inside", formatter='{b}: {c}枚'),
    levels=[
            opts.TreeMapLevelsOpts(
                treemap_itemstyle_opts=opts.TreeMapItemStyleOpts(
                    border_color="white", border_width=4, gap_width=4
                )
            ),
            opts.TreeMapLevelsOpts(
                color_saturation=[0.8, 0.5],
                treemap_itemstyle_opts=opts.TreeMapItemStyleOpts(
                    border_color="white",border_color_saturation=0.7, gap_width=4, border_width=8
                ),
            )
        ],
    )
tree.set_global_opts(
    title_opts=opts.TitleOpts(
        title="各国家不同运动项目的奖牌情况（可点击下钻到各国家获得奖牌情况）",
        subtitle = '数据来源：百度体育',
        pos_left='center', 
        title_textstyle_opts=opts.TextStyleOpts(font_size=20),
        pos_top='1%'
        ),
    legend_opts=opts.LegendOpts(
        is_show=False
        )
)
tree.render_notebook()
```

```python
tmp = df_t.query('奖牌类型 == "金牌"')['获奖时间'].value_counts().reset_index()
tmp.columns = ['日期','个数']
tmp = tmp.sort_values('日期')
for i in range(1,len(tmp)):
    tmp.iloc[i,1] = tmp.iloc[i-1,1] + tmp.iloc[i,1]
```

```python
def cal_everyday(type_):
    tmp = df_t[df_t['奖牌类型'] == type_]['获奖时间'].value_counts().reset_index()
    tmp.columns = ['日期',type_+'总数']
    tmp = tmp.sort_values('日期')
    for i in range(1,len(tmp)):
        tmp.iloc[i,1] = tmp.iloc[i-1,1] + tmp.iloc[i,1]
    return tmp
```

```python
tmp = pd.merge(cal_everyday("金牌"),pd.merge(cal_everyday("银牌"),cal_everyday("铜牌"),how='outer'),how='outer').sort_values('日期')
tmp.fillna(method='ffill',inplace=True)
tmp.fillna(0,inplace=True)
```

```python
line_style = {
    'normal': {
        'width': 5,
        'shadowColor': 'rgba(155, 18, 184, .3)',
        'shadowBlur': 10,
        'shadowOffsetY': 15,
        'shadowOffsetX': 10,
        'curve': 0.5
    }
}
```

```python
from pyecharts.commons.utils import JsCode
color_js = """
            new echarts.graphic.LinearGradient(
                                0,
                                0,
                                0,
                                1,
                                [{offset: 0, color: '#FFEDBC'},
                                 {offset: 1, color: '#ED4264'}],
                                false)
           """
```

👇接下来再看看我国每日的不同奖牌总数的变化情况吧

```python
tl = Timeline(init_opts=opts.InitOpts(theme='light',width='980px'))
tl.add_schema(is_auto_play=True, is_loop_play=True, 
                    is_timeline_show=False, play_interval=1500)
values = []
values_s = []
values_b = []
for i in tmp['日期']:
    values.append(tmp[tmp['日期'] == i]['金牌总数'].values.tolist()[0])
    values_s.append(tmp[tmp['日期'] == i]['银牌总数'].values.tolist()[0])
    values_b.append(tmp[tmp['日期'] == i]['铜牌总数'].values.tolist()[0])
    line = (
        Line(init_opts=opts.InitOpts(width='980px',height='600px',bg_color=JsCode(color_js),))
        .add_xaxis(tmp['日期'].tolist())
        .add_yaxis(
            'Gold_Sum',
            values,
            is_smooth=True,
            linestyle_opts=line_style
            )
        .add_yaxis(
            'Silver_Sum',
            values_s,
            is_smooth=True,
            linestyle_opts=line_style
            )
        .add_yaxis(
            'Bronzes_Sum',
            values_b,
            is_smooth=True,
            linestyle_opts=line_style
            )
        .set_global_opts(
             title_opts=opts.TitleOpts(
                title="我国金、银、铜牌总量日变化状况",
                subtitle='数据来源：百度体育',
                pos_right='center',
                pos_top='2%'
                ),
            legend_opts=opts.LegendOpts(
                is_show=True,
                legend_icon='circle',
                pos_right='10%',
                pos_top='7%'
                ),
            xaxis_opts=opts.AxisOpts(
                axislabel_opts={'rotate':90}
            ),
            yaxis_opts=opts.AxisOpts(
                max_=15,
                axisline_opts=opts.AxisLineOpts(
                    is_show=False
                ),
                splitline_opts=opts.SplitLineOpts(
                    is_show=True
                )
            ),
        )
    )
    tl.add(line,'{}'.format(i))
tl.render_notebook()
```
